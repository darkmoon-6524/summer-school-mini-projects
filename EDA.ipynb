```python
import pandas as pd
import seaborn as sns                      
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv('/content/student_data.csv')
df.head(10)
```

```python
df.shape
```

```python
duplicate_rows = df[df.duplicated()]
print(duplicate_rows.shape)
```

```python
df = df.drop_duplicates()
```

```python
df.count()
```

```python
df.isnull().sum()
```

```python
df = df.rename(columns = {"Student ID": "ID", "Student Name": "Name", "Field of Study": "Field"})
df.head()
```

```python
df.Field.value_counts().plot(kind='bar', figsize=(10,5))
plt.title("No of students in each field")
plt.ylabel("No of students")
plt.xlabel("Field")
```

```python
plt.figure(figsize = (10,6))
sns.scatterplot(x = df['Current Semester'], y = df['Fees'])
```

```python
plt.figure(figsize = (10,6))
c= df.corr()
sns.heatmap(c,annot=True)
```
